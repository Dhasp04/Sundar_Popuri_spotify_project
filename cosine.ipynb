# Forming Cosine Similarity Groups

Cosine Similarity groups are found to form as a base for classification algorithms such as logistic regression, random forrests and XGboost. 

## Loading Data

```python
import pandas as pd
df = pd.read_csv("data/spotify_data.csv", index_col= None)
df.head()
```

```python
data = df.drop(columns = ["track_name"])
data.info()
```

### Cosine Similarity

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python

## Getting numerical features for cosine similarity 
features = data.iloc[:, 3:].drop(columns = "track_genre")

features.head()

```

```python
# making track_genre numerical 
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=True, handle_unknown="ignore")
genre_matrix = encoder.fit_transform(df[["track_genre"]])

genre_df = pd.DataFrame.sparse.from_spmatrix(
    genre_matrix,
    index=features.index,
    columns=encoder.get_feature_names_out(["track_genre"])
)

data_cosine = pd.concat([features, genre_df], axis=1)
data_cosine.head()
```

```python
#scaling continous columns
from sklearn.preprocessing import StandardScaler

audio_cols = [
    "popularity", "danceability", "energy", "loudness", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "duration_s"
]

scaler = StandardScaler()
data_cosine[audio_cols] = scaler.fit_transform(data_cosine[audio_cols])

data_cosine.head()
```

```python
from sklearn.neighbors import NearestNeighbors
from scipy.sparse import csr_matrix

X = csr_matrix(data_cosine.values)

nn_model = NearestNeighbors(
    metric = "cosine", algorithm = 'brute'
    )

nn_model.fit(X)
```

```python
np.random.seed(42)

all_tracks = data["track_id"].unique()
negative_rows = []

positive_map = (
    positive_pairs.groupby("seed_track_id")["similar_track_id"]
    .apply(set)
    .to_dict()
)

for seed_id, pos_set in positive_map.items():
    excluded = pos_set | {seed_id}
    n_neg = len(pos_set)

    sampled = []
    sampled_set = set()

    while len(sampled) < n_neg:
        candidate = np.random.choice(all_tracks)
        if candidate not in excluded and candidate not in sampled_set:
            sampled.append(candidate)
            sampled_set.add(candidate)

    negative_rows.extend(
        {
            "seed_track_id": seed_id,
            "similar_track_id": neg_id,
            "target": 0
        }
        for neg_id in sampled
    )

neg_pairs = pd.DataFrame(negative_rows)
```

```python
neg_pairs.head()
```

### Creating Pairwise Pairs 

```python
pairwise_pairs = pd.concat([positive_pairs, neg_pairs], ignore_index=True)

feature_cols = [
    "popularity", "danceability", "energy", "loudness", "speechiness",
    "acousticness", "instrumentalness", "liveness", "valence", "tempo",
    "duration_s",
]

song_features = data.set_index("track_id")[feature_cols]

seed_features = (
    song_features.loc[pairwise_pairs["seed_track_id"]]
    .reset_index(drop=True)
)

candidate_features = (
    song_features.loc[pairwise_pairs["similar_track_id"]]
    .reset_index(drop=True)
)

pairwise_X = seed_features.sub(candidate_features, axis=0).abs()
pairwise_X.columns = [f"{col}_diff" for col in feature_cols]


pairwise_data = pd.concat(
    [pairwise_X, pairwise_pairs[["target"]].reset_index(drop=True)],
    axis=1
)
pairwise_data.head()
```

```python
lookup = data.set_index("track_id")

pairwise_data_with_ids = pairwise_pairs.copy()

pairwise_data_with_ids["same_genre"] = (
    lookup.loc[pairwise_data_with_ids["seed_track_id"], "track_genre"].values
    ==
    lookup.loc[pairwise_data_with_ids["similar_track_id"], "track_genre"].values
).astype(int)

pairwise_data_with_ids["same_artist"] = (
    lookup.loc[pairwise_data_with_ids["seed_track_id"], "artists"].values
    ==
    lookup.loc[pairwise_data_with_ids["similar_track_id"], "artists"].values
).astype(int)

pairwise_data_with_ids["same_album"] = (
    lookup.loc[pairwise_data_with_ids["seed_track_id"], "album_name"].values
    ==
    lookup.loc[pairwise_data_with_ids["similar_track_id"], "album_name"].values
).astype(int)

pairwise_final = pd.concat(
    [
        pairwise_data_with_ids[["seed_track_id", "similar_track_id"]],
        pairwise_X.reset_index(drop=True),
        pairwise_data_with_ids[["same_genre", "same_artist", "same_album", "target"]].reset_index(drop=True)
    ],
    axis=1
)

pairwise_final.head()
```

```python
pairwise_final.to_csv("data/pairwise_data.csv", index = False)
```
